##Ερώτημα 2: Υπολογισμός παραγώγων

###2.1) $f(x) = x^{Τ}Ax + b^{Τ}x$, να υπολογίσετε την παράγωγο της f ως προς το διάνυσμα x ∈ ℝ^{d}

Έχουμε :

$f(x) = x^{T}Ax + b^{T}x,    A=A^{T} $ και $ b \in R^{d} $ \\

$$
\frac{\partial f}{\partial x} = \frac{\partial x^{T}Ax + b^{T}x}{\partial x} = \frac{\partial x^{T}Ax}{\partial x} + \frac{\partial b^{T}x}{\partial x}
$$

Από το matrixcookbook που μας δόθηκε και τις σχέσεις (69) και (81) έχουμε ότι

$
\frac{\partial x^{T}Ax}{\partial x} = x(A+A^{T}) = 2Ax \\
\frac{\partial b^{T}x}{\partial x} = b\\
$

Άρα :
$$
\frac{\partial f}{\partial x} = 2Ax + b
$$


###2.2) Έστω Α, Β, Χ πίνακες πραγματικών αριθμών διάστασης d x d με τάξη (rank) ίση με d. Nα βρείτε, αναλυτικά και σε κλειστή μορφή (closed form solution) το ολικό ελάχιστο του παρακάτω προβλήματος βελτιστοποίησης : $$\begin{equation}
\mathop{min}_{\textbf{x}}
\end{equation} ||A-XB||^{2}_{F}$$

Έστω $$f(X) = ||A-XB||^{2}_F$$ \\

Έχουμε :
$$∇_X f(X) = \frac{d||A-XB||^{2}_F}{dX} = \frac{\partial Tr((A-XB)(A-XB)^{H})}{\partial X} = 2(A-XB)$$

Η δεύτερη παράγωγος είναι 2 > 0, άρα το σημείο στο οποίο μηδενίζεται η παράγωγος είναι ολικό ελάχιστο. Άρα:

$$∇_X f(X) = 0 ⇒ \frac{d||A-XB||^{2}_F}{dX} = 0 ⇒ 2(A-XB) = 0 ⇒ A-XB = 0 ⇒ A = XB ⇒ X = \frac{A}{B} ⇒ X = AB^{-1}$$

##Ερώτημα 3: Gradient descent

###Για την $f_{1} = (x_1 - 2)^{2} + (x_2 - 3)^{2} $, έχουμε μερικές παραγώγους
$$
   \frac{\partial f_1}{\partial x1} = 2(x_1-2) \\
   \frac{\partial f_1}{\partial x2} = 2(x_2-3)
$$

####Ξεκινάμε τον αλγόριθμο με αρχικές τιμές $x_1 = x_2 = 0$, iterations = 10 και learning rate = 0.5

```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x=0, x2=0, iterations = 10, learning_rate = 0.5, convergence = 0):

  def f1(x1,x2):
    return (x1-2)**2 + (x2-3)**2

  def dfx(x):
    return (2*(x-2))

  def dfx2(x2):
    return (2*(x2-3))

  curX = x # The algorithm starts at x=1
  curX2 = x2 # x2=0
  rate = learning_rate # Learning rate
  precision = 0.0001 #This tells us when to stop the algorithm
  previous_step_size = 1 #
  maxIters = iterations # maximum number of iterations
  iters = 0 #iteration counter

  df = np.array([])
  f = np.array([])

  for aIter in range(maxIters):
    #x1
    prevX=curX
    curX=curX-rate*dfx(curX)
    diff=abs(curX-prevX)

    #x2
    prevX2=curX2
    curX2 = curX2-rate*dfx2(curX2)
    diff2 = abs(curX2-prevX2)

    df = np.append(df, (diff + diff2))  # times paragwgoy
    f = np.append(f, f1(curX, curX2))   # times sunarthshs

    if(convergence == 0):
      if diff<=precision or diff2<=precision:
        print("Iteration ", aIter)
        print("The local minimum occurs at x1", curX, " x2 ", curX2)
        print("df = ", df[aIter])
        print("f = ", f[aIter])
        break
    else:
      if curX == prevX and curX2 == prevX2:
        print("Iterations for convergence ", aIter)
        break

    print("df = ", df[aIter])
    print("f = ", f[aIter])
    print("Iteration",aIter,"\nX1 value is",curX) #Print iterations
    print("X2 value is",curX2)


  iterations = np.array(range(0,aIter+1))
  plt.figure(figsize = (8,6))
  plt.plot(iterations, f, color='blue',markerfacecolor='red', markersize=10, linestyle='dashed')
  plt.xlabel("iterations")
  plt.ylabel("f")
  plt.show()

gradient_descent()  # Iteration  1 and local minimum at (x1 = 2.0, x2 = 3.0)
```

####Και τώρα διατρέχουμε για διαφορετικές τιμές του learning rate και του πλήθους επαναλήψεων

```python
gradient_descent(0, 0, 1000, 0.01)    # Iterations  297 and local minimum at (x1 = 1.9951426370929788, x2 = 2.992713955639469)

gradient_descent(0, 0, 1000, 0.001)   # Iterations 1843 and local minimum at (x1 = 1.9501404758253953, x2 = 2.925210713738091)

gradient_descent(0, 0, 10000, 0.0001) # Iterations 6931 and local minimum at (x1 = 1.5001221337650217, x2 = 2.250183200647539)
```

###Για την $f_{2} = (1 - (x_2-3))^{2} + 20((x_1 + 3)-(x_2 - 3))^{2} $, έχουμε μερικές παραγώγους
$$
   \frac{\partial f_2}{\partial x1} = 40(x_1+3) - 40(x_2-3)^{2} \\
   \frac{\partial f_2}{\partial x2} = -2 + 2(x_2 - 3)-80(x_1 + 3)(x_2 - 3) + 80(x_2 - 3)^{3}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def gradiant_descent2(x1=0,x2=0,iterations2=10, learning_rate2=0.5, convergence=0):

  def dfx(x1, x2):
    return (40*(x1+3)-40*(x2-3)**2)

  def dfx2(x1, x2):
    return -2 + 2*(x2-3) - 80*(x1+3)*(x2-3) + 80*(x2-3)**3

  def f2(x,x2):
    return (1-(x2-3))**2 + 20*((x+3)-(x2 - 3)**2)**2

  curX = x1 # The algorithm starts at x1=0
  curX2 = x2 # x2=0
  rate = learning_rate2 # Learning rate
  precision = 0.0001 #This tells us when to stop the algorithm
  previous_step_size = 1 #
  maxIters = iterations2 # maximum number of iterations
  iters = 0 #iteration counter

  df = np.array([])
  f = np.array([])

  for aIter in range(maxIters):
    #x1
    prevX=curX
    curX=curX-rate*dfx(curX, curX2)
    diff=abs(curX-prevX)

    #x2
    prevX2=curX2
    curX2 = curX2-rate*dfx2(prevX, curX2)
    diff2 = abs(curX2-prevX2)

    df = np.append(df, (diff + diff2))
    f = np.append(f, f2(curX, curX2))

    print("df = ", df[aIter])
    print("f = ", f[aIter])
    print("Iteration",aIter,"\nX1 value is",curX) #Print iterations
    print("X2 value is",curX2) #Print iterations

    if(convergence == 0):
      if diff<=precision and diff2<=precision:
        print("Iteration ", aIter)
        print("The local minimum occurs at x1", curX, " x2 ", curX2)
        print("df = ", df[aIter])
        print("f = ", f[aIter])
        break
    else:
      if curX == prevX and curX2 == prevX2:
        print("Iterations for convergence ", aIter)
        break

  iterations = np.array(range(0,aIter+1))
  plt.figure(figsize = (8,6))
  plt.plot(iterations, f, color='blue',markerfacecolor='red', markersize=10, linestyle='dashed')
  plt.xlabel("iterations")
  plt.ylabel("f2")
  plt.show()

#gradiant_descent2()  # error
```
